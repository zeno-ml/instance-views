```python
import zeno
from zeno import ZenoParameters
import pandas as pd
```

```python
df = pd.DataFrame({
    "data": [
        [
            {"role": "assistant", "content": "# Hello \n * first bullet \n * second bullet"},
            {"role": "user", "content": "# Hello \n * first bullet \n * second bullet"},
         ],
    ],
    "label": [
        "## Hello \n * first bullet \n * second bullet"
        ]
})
```

```python
zeno.zeno(ZenoParameters(metadata=df, data_column="data", label_column="label", view="openai-chat-markdown/index.mjs"))
```
